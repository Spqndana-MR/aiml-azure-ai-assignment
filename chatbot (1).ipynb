```python
# -----------------------------
# 4. INTERACTIVE CHATBOT
# -----------------------------
print("\n🤖 Chatbot is ready! Type 'exit' to stop.\n")
print("Here are some example questions you can ask:")
print("- I have a fever")
print("- I have a cold")
print("- I have a headache")
print("- What is the weather like today?")
print("- Tell me a story.")
print("\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Bot: Goodbye! 👋")
        break

    # -------- BERT (understanding) --------
    sentiment = bert(user_input)[0]['label']
    print(f"(Debug: User sentiment detected as: {sentiment})")

    # -------- RULE-BASED FIX (IMPORTANT) --------
    if "fever" in user_input.lower():
        print("Bot: You may have a fever. Please take rest and consult a doctor.")
        continue

    elif "cold" in user_input.lower():
        print("Bot: It seems like a cold. Stay warm and drink fluids.")
        continue

    elif "headache" in user_input.lower():
        print("Bot: You might have a headache. Take rest and avoid screen time.")
        continue

    # -------- GPT (controlled output) --------
    response = gpt(
        "User: " + user_input + "\nBot:",
        max_new_tokens=30,
        do_sample=True,
        temperature=0.7
    )

    text = response[0]['generated_text']

    # Clean output (important)
    # GPT might generate the prompt itself, so we only take the part after 'Bot:'
    reply = text.split("Bot:")[-1]

    print("Bot:", reply.strip())
```
